```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train=pd.read_csv('train (3).csv')
test=pd.read_csv('test (1).csv')
```

```python
train.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
a=le.fit_transform(train['HomeTeam'])
b=le.fit_transform(train['AwayTeam'])
```

```python
a=pd.DataFrame(a,columns=['home'])
b=pd.DataFrame(b,columns=['Away'])
```

```python
train=train.join(a)

```

```python
train=train.join(b)
```

```python
train.info()
```

```python
train
```

```python
train.fillna(train.mean(),inplace=True)
```

```python
train['Date'] = pd.to_datetime(train['Date'])
train['year']=train['Date'].apply(lambda time:time.year)
train['Month'] = train['Date'].apply(lambda time: time.month)
train['Day of Week'] = train['Date'].apply(lambda time: time.dayofweek)
```

```python
train.head()
```

```python
X=train[[ 'B365H', 'B365D', 'B365A', 'BWH', 'BWD', 'BWA',
       'IWH', 'IWD', 'IWA', 'LBH', 'LBD', 'LBA','WHH', 'WHD',
       'WHA', 'VCH',
       'VCD', 'VCA', 'home', 'Away','year','Month','Day of Week']]
y=train['FTR']
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
cl=LogisticRegression()
```

```python
cl.fit(X_train,y_train)
```

```python
pred=cl.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm=confusion_matrix(y_test,pred)
```

```python
cm
```

```python
pred
```

```python
cr=classification_report(y_test,pred)
```
